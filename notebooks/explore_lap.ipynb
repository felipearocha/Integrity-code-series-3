# ICS2 Week 3 — F1 Lap Simulation
Interactive exploration of simulation outputs.

```python
import sys
sys.path.insert(0, '../src')
from simulation.physics_model import CarParams, TrackParams, run_lap
import matplotlib.pyplot as plt
import numpy as np
```

```python
cp = CarParams()
tp = TrackParams()
res = run_lap(cp, tp, verbose=True)
print('Lap time: %.2f s' % res['lap_time'])
```

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)
axes[0].plot(res['s'], res['v_kmh'], color='steelblue')
axes[0].set_ylabel('Speed km/h')
axes[1].plot(res['s'], res['T_tire'], color='orange')
axes[1].axhline(95, color='red', ls='--', alpha=0.7)
axes[1].set_ylabel('Tyre Temp C')
axes[2].plot(res['s'], res['SOC']*100, color='green')
axes[2].axhline(35, color='red', ls='--', alpha=0.7)
axes[2].set_ylabel('SOC %')
axes[2].set_xlabel('Arc length m')
plt.tight_layout()
plt.show()
```
